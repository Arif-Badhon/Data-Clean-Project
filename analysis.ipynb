```python
import pandas as pd
from ydata_profiling import ProfileReport 
import re
from wordcloud import WordCloud
import matplotlib
matplotlib.use('TkAgg')
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Data/archive/Raw_Dataset.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
profile = ProfileReport(df, title='Summary Report')
profile.to_notebook_iframe()
```

```python
df['Salary'] = df['Salary'].str.replace('£', '').str.replace('K', '000')
df.head()
```

```python
df['Salary'] = df['Salary'].str.replace(r'\(Employer est.\)', '', regex=True).str.replace(r'\(Glassdoor est.\)', '', regex=True)
df.head()
```

```python
def convert_to_year(salary):
    if 'Per Hour' in str(salary):
        hourly_rate = re.findall(r'\d+\.?\d*', salary)
        if hourly_rate:
            return float(hourly_rate[0])*2080
    return salary

df['Salary'] = df['Salary'].apply(convert_to_year)
df.head()
```

```python
df['Salary'] = df['Salary'].astype(str)
df[['salary_min', 'salary_max']] = df['Salary'].str.split('-', expand=True)
df.head()
```

```python
df['salary_max'] = df['salary_max'].str.strip()
df['salary_min'] = df['salary_min'].str.strip()
df.head()
```

```python
df['salary_max'] = pd.to_numeric(df['salary_max'], errors='coerce')
df['salary_min'] = pd.to_numeric(df['salary_min'], errors='coerce')
df.head()
```

```python
df['Location'] = df['Location'].astype(str)
df[['city', 'country']] = df['Location'].str.split(',', n=1, expand=True)
df.head()
```

```python
skills_split = df['Skills'].str.get_dummies(sep=', ')
skills_split
```

```python
df_with_skill = pd.concat([df, skills_split], axis=1)
df_with_skill.head()
```

```python
skill_frequency = df_with_skill.iloc[:, df_with_skill.columns.get_loc('AI'):].sum()
skill_frequency
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(skill_frequency)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Popular Skill', fontsize=16)
plt.figure(figsize=(40, 30))
plt.show(block=True)
input("Press Enter to close")
plt.close()
```

```python
skill_columns = df_with_skill.iloc[:, df_with_skill.columns.get_loc('AI'):].sum().sort_values(ascending=False)
top_10_skills = skill_columns.head(10)
```

```python
plt.figure(figsize=(10, 6))
top_10_skills.plot(kind='bar', color='green')
plt.title('Top 10 most popular skills', fontsize=16)
plt.gca().invert_yaxis()
plt.show()
```
